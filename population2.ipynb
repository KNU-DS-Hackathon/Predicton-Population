```python
import pandas as pd
import glob

# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
def getDF(year, file_path):
    for fName in glob.glob(file_path):
        df = pd.read_csv(fName, encoding='cp949', thousands=',')
        ret = df[list(df)[0:1]]

        ret['행정코드'] = ret['행정구역'].str.split('(').str[-1].str.replace(')', '',).str.strip()
        ret['행정구역'] = ret['행정구역'].str.split('(').str[0].str.strip()
        ret['인구수(명)'] = df[list(df)[4:20]].sum(axis=1)
        ret.insert(0, '연도', year)

        return ret
```

```python


result = pd.DataFrame()

for year in range(2008, 2023):
    path = './rawData/' + str(year) + '12' + '*.csv'
    new_df = getDF(year, path)
    result = pd.concat([result, new_df])
        
result.sort_values(by=['행정코드', '연도'], ascending=True, inplace=True)

result.to_csv('result_popul2.csv', index=False)
print(result.shape[0])
# print(result.head(100))
# print(new_df.dtypes)
```
