```python

from statsmodels.tsa.stattools import adfuller
import numpy as np

# Preprocessing steps: Differencing and taking the log
def preprocess_data(group):
    # Applying log transformation
    group = np.log(group + 1)  # Adding 1 to handle zero values
    # Differencing
    group_diff = group.diff().dropna()
    return group_diff


# Group by '행정구역' and apply the preprocessing
grouped_data = data.drop(['행정구역(시도)',	'행정구역(시군구)'], axis=1).groupby('행정구역').apply(lambda x: preprocess_data(x.sort_values('연도').set_index('연도')))
original_grouped_data = data[data['연도']==2010].drop(['행정구역(시도)','행정구역(시군구)'], axis=1)
# Display the first few rows of the preprocessed data


grouped_data = grouped_data.reset_index()

grouped_data.head(13)
```

```python
from statsmodels.tsa.stattools import adfuller

# 가정: 'your_time_series'는 분석하고자 하는 시계열 데이터임

for col in df_adf.columns:

    adf_test_result = adfuller(df_adf[col])

    print('ADF Statistic: %f' % adf_test_result[0])
    print('p-value: %f' % adf_test_result[1])
    print('Critical Values:')
    for key, value in adf_test_result[4].items():
        print('\t%s: %.3f' % (key, value))

    # 결과 해석
    if adf_test_result[0] < adf_test_result[4]["5%"]:
        print("시계열 데이터는 정상성을 가짐 (귀무 가설 기각)")
    else:
        print("시계열 데이터는 비정상성 (귀무 가설 채택)")
```

```python
def forecasting_features_var(data, steps=10):
    from statsmodels.tsa.api import VAR
    
    # 전달받은 데이터에 object 타입이 있는지 확인
    #print(data.dtypes)
    
    # Assuming data is a DataFrame with all necessary columns for VAR model
    # VAR 모델은 시계열 데이터의 정상성을 요구하므로 필요한 경우 데이터 변환 수행
    # VAR 모델 피팅
    model = VAR(data)
    results = model.fit()

    # 예측
    forecast = results.forecast(data.values[-results.k_ar:], steps=steps)

    # 연도의 마지막 날짜를 구하는 함수
    def get_year_end(year):
        return pd.Timestamp(year=year, month=12, day=31)

    # 데이터의 마지막 연도
    last_year = data.index[-1].year

    # 예측 결과를 DataFrame으로 변환
    future_dates = pd.date_range(get_year_end(last_year), periods=steps, freq='Y')
    forecast_df = pd.DataFrame(forecast, index=future_dates, columns=data.columns)


    return forecast_df
```

```python
#grouped_data = grouped_data.reset_index()
print(grouped_data.head())

for index in grouped_data['행정구역'].unique():
    # '행정구역'을 제외하고 모델에 데이터 전달
    data = grouped_data[grouped_data['행정구역'] == index].set_index('연도').drop('행정구역', axis=1)
    data.index = pd.to_datetime(data.index, format='%Y')  # 인덱스를 datetime으로 변환

    #print(data.head())
    forecast_df = forecasting_features_var(data, steps=40)

    # 예측 결과에 '행정구역' 정보 추가
    forecast_df['행정구역'] = index
    
    # forecast_df의 인덱스에서 연도만 추출
    forecast_df.index = forecast_df.index.year + 1
    
    forecast_df.reset_index(inplace=True)
    forecast_df.rename(columns={'index': '연도'}, inplace=True)
    #forecast_df.set_index(['행정구역'], inplace=True)
    

    # 예측 결과를 원본 데이터에 추가
    grouped_data = pd.concat([grouped_data, forecast_df])
```

```python
# 로그 변환 및 차분 역변환 함수
def reverse_preprocess(group_diff, original_log_value):
    reversed_df_list = []

    for column in group_diff.columns:
        # 차분 역변환 (첫 번째 값 설정 및 누적합)
        first_log_value = original_log_value[column]
        reversed_diff = np.r_[first_log_value, group_diff[column]].cumsum()

        # 로그 변환 역변환 (지수 함수 적용)
        reversed_log = np.exp(reversed_diff) - 1

        # 역변환된 데이터를 DataFrame으로 변환
        reversed_df = pd.DataFrame(reversed_log, columns=[column])
        reversed_df_list.append(reversed_df)

    # 모든 열을 결합
    reversed_df_full = pd.concat(reversed_df_list, axis=1)
    return reversed_df_full

original_log_values = np.log(original_grouped_data.set_index('행정구역') + 1)

# 각 행정구역에 대해 역변환 적용
reversed_data = pd.DataFrame()
for region in grouped_data['행정구역'].unique():
    group_diff = grouped_data[grouped_data['행정구역'] == region].sort_values('연도')
    original_log_value = original_log_values.loc[region]

    reversed_group = reverse_preprocess(group_diff.drop(['행정구역', '연도'], axis=1), original_log_value)
    reversed_group['행정구역'] = region

    # 연도 데이터 할당
    years_with_2010 = np.insert(group_diff['연도'].values, 0, 2010)
    reversed_group['연도'] = years_with_2010

    # 모든 행정구역의 역변환 데이터를 하나의 데이터프레임에 추가
    reversed_data = pd.concat([reversed_data, reversed_group], ignore_index=True)

# 결과 확인
print(reversed_data.head(15))
```

```python
# MAPE 계산 함수
def calculate_mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```
