```python
import pandas as pd
```

```python
def function(file_path):
    
    df = pd.read_csv(file_path, low_memory=False, encoding='cp949')
    
    # string -> float
    df['총인구(명)'] = df['총인구(명)'].replace('X', 0).astype(str)
    df['총인구(명)'] = df['총인구(명)'].astype(float)
    
    # 나이 정의
    ages = []
    for i in range(6,22):
        ages.append(' '+str(i)+'세')
        
    # 6세 ~ 21세 빼고 삭제
    df = df[df['연령별'].isin(ages)]
    
    # 쓸데 없는 column 삭제
    df = df.drop(['C연령별','내국인_성비', '내국인_여자(명)','내국인_남자(명)','내국인(명)','총인구_성비','총인구_여자(명)','총인구_남자(명)'], axis=1)
    
    #행정구역별(읍면동)에 따라 6세~21세 총인구 종합
    new_df = df.groupby(['C행정구역별(읍면동)','행정구역별(읍면동)','시점'])['총인구(명)'].sum().reset_index()
    
    
    return new_df
```

```python
file_path = []
df = []
for i in range(2015,2022):
    file_path.append('./Data/'+str(i)+'.csv')
    
for path in file_path:
    df.append(function(path))
```

```python
# 병합

concat_df = pd.concat(df)

concat_df = concat_df.sort_values(by=['C행정구역별(읍면동)','시점'])

concat_df.head(30)
```

```python
# csv로 저장

concat_df.to_csv('2015-2021 전국 읍면동 학령인구 데이터.csv', index=False)
```
